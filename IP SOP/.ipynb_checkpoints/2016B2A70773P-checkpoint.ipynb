```python
# Importing all the necessary packages:
import matplotlib.pyplot as plt
from skimage import data
from skimage.feature import blob_log
import numpy as np
import cv2
```

### Reading Image from file and Gray Scaling

The given may/maynot be in the gray scale format, hence we read the file and convert it to desired gray scale format. 

```python
# Reading the image 'Input.png' into variable 'img'
# opening in mode = 0 (Gray Scale): 
img = cv2.imread('Input.png',0)

# Plotting the image on console:
plt.imshow(img,cmap ='gray')
```

```python
# Converting 'img' into numpy array for easier calculation
img = np.asarray(img)
```

### Binary Conversion of Image

We calculate the threshold using Otsu Threshold, and apply it on the image. 

```python
# Thresholding image using Otsu Threshold:
ret2,threshold = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Plotting the binary resultant image on console:
plt.imshow(threshold, cmap ='gray') 
```

### Canny Edge Detection

Using the binary image usr use Canny Edge Detection technique to fetch the edge boundaries in the image.

```python
# Using Canny Detection in OpenCV
edges = cv2.Canny(threshold,5.5,0.001)

# Plotting the edge detected image: 
plt.imshow(edges,cmap = 'gray')
```

### Smoothing The Edges

We now try to use GaussianBlur to connect and smoothen the edge for better hole filling.

```python
# Smoothing of edges for better hole filling
smooth = cv2.GaussianBlur(edges,(3,3),0)

# Plotting the hole filled image: 
plt.imshow(smooth,cmap = 'gray')
```

### Filling Holes

The vehicle holes by canny edge detection are filled as 255 value. This is achieved using morphological close operation. 

```python
# Defining the kernal for the Morphological Close Operation:
kernel = np.ones((9,9),np.uint8)
closing = cv2.morphologyEx(smooth, cv2.MORPH_CLOSE, kernel)


# Plotting the hole filled image: 
plt.imshow(closing,cmap = 'gray')
```

### Cropping the image

The resultant image is cropped to the area containing the maximum blobs.

```python
# Cropping the image:
crop_img = closing[175:250,5:475]

# Thresholding again using Open-CV for binary image: 
ret, thres = cv2.threshold(crop_img,20,255,0)

# Plotting the cropped image: 
plt.imshow(thres,cmap = 'gray')
```

### Blob Analysis

Using blob analysis we try to estimate the vehicles.

```python
# Converting thres (gray scale) to RGB image but black and white:
finalimg = cv2.cvtColor(thres,cv2.COLOR_GRAY2RGB)

# Finding contours using the binary cropped image: 
_,contours,_ = cv2.findContours(thres, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Adding contours to the image:
final = cv2.drawContours(finalimg, contours, -1, (255, 0, 0), 2)

# Plotting the image with contours: 
plt.imshow(final)
```
